```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('clean_application_train.csv')
```

```python
df.head()
```

```python
df['LOAN_TO_INCOME'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
```

```python
df['ANNUITY_TO_INCOME'] = df['AMT_ANNUITY'] / df['AMT_INCOME_TOTAL']
```

```python
df['EMPLOYMENT_YEARS'] = abs(df['DAYS_EMPLOYED']) / 365
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df[['LOAN_TO_INCOME', 'TARGET']].groupby('TARGET').mean()
```

```python
df.isnull().sum().sum()
```

```python
df.to_csv("fe_application_train.csv", index=False)
```
