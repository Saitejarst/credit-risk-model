```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv("application_train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
missing  = df.isnull().mean() * 100
missing = missing.sort_values(ascending = False)
missing.head(15)
```

```python
high_missing_cols = missing[missing > 40].index
len(high_missing_cols)
```

```python
df.drop(columns = high_missing_cols , inplace = True)
```

```python
num_cols = df.select_dtypes(include = ['int64' , 'float64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_col = df.select_dtypes(include = ['object']).columns

for col in cat_col:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['AMT_INCOME_TOTAL'] = df['AMT_INCOME_TOTAL'].clip(
    lower=df['AMT_INCOME_TOTAL'].quantile(0.01),
    upper=df['AMT_INCOME_TOTAL'].quantile(0.99)
)
```

```python
df['AMT_CREDIT'] = df['AMT_CREDIT'].clip(
    lower=df['AMT_CREDIT'].quantile(0.01),
    upper=df['AMT_CREDIT'].quantile(0.99)
)
```
