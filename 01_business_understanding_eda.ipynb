```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("default")
```

```python
df = pd.read_csv("application_train.csv")
```

```python
sns.boxplot(x = 'TARGET' , y = 'AMT_INCOME_TOTAL' , data=df)
plt.title("Income vs Loan Default")
plt.show()
```

```python
df['AMT_INCOME_TOTAL'].skew()
```

```python
sns.boxplot(
    x = 'TARGET',
    y = 'AMT_INCOME_TOTAL',
    data = df[df['AMT_INCOME_TOTAL'] < df['AMT_INCOME_TOTAL'].quantile(0.99)]
)
plt.title("Income vs Default(After removing Top 1% Outliers)")
plt.show()
```

```python
sns.boxplot(x = 'TARGET' , y = 'AMT_CREDIT', data = df)
plt.title("Loan Amount vs Loan Default")
plt.show()
```
